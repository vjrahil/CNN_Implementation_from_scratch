```python
import numpy as np
from torchvision import datasets
import torchvision.transforms as transforms
import torch
import matplotlib.pyplot as plt
import pickle
```

```python
def Convolution(image,filt,s=1):
    
    (n_channel,height,_) = image.shape
    (n_filt,n_channel_filt,filt_height,_) = filt.shape
    
    out_dim = int((height - filt_height)/s) + 1
    
    out = np.zeros((n_filt,out_dim,out_dim))
    
    for cur_f in range(n_filt):
        cur_y = 0 
        out_y = 0
        while(cur_y + filt_height <= height):
            cur_x = 0
            out_x = 0
            while(cur_x + filt_height <= height):
                out[cur_f,out_y,out_x] = np.sum(filt[cur_f]*image[:,cur_y:cur_y + filt_height,cur_x:cur_x + filt_height] )
                out_x += 1
                cur_x += s
            cur_y += s
            out_y += 1
            
    return out

def MaxPooling(image,filt_height = 2,s = 2):
    (n_channel, height, width) = image.shape
    h_out = int((height-filt_height)/s) + 1
    w_out = int((width - filt_height)/s) + 1
    
    pooled = np.zeros((n_channel,h_out,w_out))
    
    for ch in range(n_channel):
        cur_y = 0 
        out_y = 0
        while(cur_y + filt_height <= height):
            cur_x = 0
            out_x = 0
            while(cur_x + filt_height <= width):
                pooled[ch,out_y,out_x] = np.max(image[ch,cur_y:cur_y + filt_height,cur_x:cur_x + filt_height])
                cur_x += s
                out_x += 1
            cur_y += s
            out_y += 1
    
    return pooled

def Flatten(input_):
    (channel,height,width) = input_.shape
    flat = input_.reshape((channel*height*width,1))
    return flat
```

```python
def Softmax(input_):
    exp = np.exp(input_)
    return exp/np.sum(exp)

def CrossEntropyLoss(output,target):
    return -np.sum(target*np.log(output))
```

```python
def conv_backprop(prev_l,current,filt,s):
    (n_filt,n_channel_filt,height_filt,_) = filt.shape
    (n_channel,height,_) = current.shape
    
    out = np.zeros(current.shape)
    out_filt = np.zeros(filt.shape)
    
    for cur_f in range(n_filt):
        cur_y = 0
        out_y = 0
        while(cur_y+height_filt<=height):
            cur_x = 0
            out_x = 0
            while(cur_x+height_filt<=height):
                out_filt[cur_f] += prev_l[cur_f,out_y,out_x] * current[:,cur_y:cur_y+height_filt,cur_x:cur_x+height_filt]
                out[:,cur_y:cur_y+height_filt,cur_x:cur_x+height_filt] += prev_l[cur_f,out_y,out_x] * filt[cur_f]
                cur_x += s
                out_x += 1
            cur_y += s
            out_y += 1
    return out_filt,out

#return a tuple
def return_index(arr):
    idx = np.nanargmax(arr)
    pair = np.unravel_index(idx,arr.shape)
    return pair

def pool_backprop(prev_l,origin,height_filt,s):
    
    (n_channel,height,_) = origin.shape
    
    out = np.zeros(origin.shape)
    
    for ch in range(n_channel):
        cur_y = 0
        out_y = 0
        while(cur_y + height_filt <= height):
            cur_x = 0
            out_x = 0
            while(cur_x + height_filt <= height):
                (a,b) = return_index(origin[ch,cur_y:cur_y + height_filt,cur_x:cur_x + height_filt])
                out[ch,cur_y + a,cur_x + b] += prev_l[ch,out_y,out_x]
                out_x += 1
                cur_x += s
            out_y += 1
            cur_y += s
        return out
        
```

```python
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])

train_data = datasets.CIFAR10('Data',train = True,download = True,transform = transform)
test_data = datasets.CIFAR10('Data',train = False,download = True,transform = transform)
```

```python
def imshow(image):
    image = image/2 + 0.5 #un- normalize it
    plt.imshow(np.transpose(image,(1,2,0)))
    
def InitializeFilter(size,scale = 1.0):
    stddev = scale/np.sqrt(np.prod(size))
    return np.random.normal(loc = 0,scale = stddev, size = size)

def InitializeWeight(size):
    return np.random.normal(size = size) * 0.01
```

```python
batch_size = 20
num_worker = 0
trainloader = torch.utils.data.DataLoader(train_data,batch_size = batch_size,num_workers = num_worker)
testloader = torch.utils.data.DataLoader(test_data,batch_size = batch_size,num_workers = num_worker)
```

```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
def ForwardPass(image,param,conv_s,pool_f,pool_s):
    
    [filt1,filt2,w3,w4] = param
    conv_1 = Convolution(image,filt1,conv_s)
    conv_1[conv_1<=0] = 0
    
    conv_2 = Convolution(conv_1,filt2,conv_s)
    conv_2[conv_2<=0] = 0
    pooled = MaxPooling(conv_2,pool_f,pool_s)
    fc = Flatten(pooled)
    layer_1 = w3.dot(fc)
    layer_1[layer_1<=0] = 0
    
    layer_2 = w4.dot(layer_1)
    
    out = Softmax(layer_2)
    
    out_layers = [conv_1,conv_2,fc,layer_1,layer_2]
    return out,out_layers
```

```python
def Implementation(image,label,param,conv_s,pool_f,pool_s):
    
    out,out_layers = ForwardPass(image,param,conv_s,pool_f,pool_s)
    
    [conv_1,conv_2,fc,layer_1,layer_2] = out_layers
    [filt1,filt2,w3,w4] = param
    
    #FORWARD PASS
    
#     conv_1 = Convolution(image,filt1,conv_s)
#     conv_1[conv_1<=0] = 0
    
#     conv_2 = Convolution(conv_1,filt2,conv_s)
#     conv_2[conv_2<=0] = 0
    
#     pooled = MaxPooling(conv_2,pool_f,pool_s)
    
#     #(n_filt,dim,_) = pooled.shape
#     #pooled.reshape((n_filt*dim*dim,1))
#     fc = Flatten(pooled)
    
#     layer_1 = w3.dot(fc)
#     layer_1[layer_1<=0] = 0
    
#     layer_2 = w4.dot(layer_1)
    
#     out = Softmax(layer_2)
    
    #LOSS
    
    loss = CrossEntropyLoss(out,label)
    
    #BACKPROPOGATION
    
    final_error = out - label 
    
    dw4 = final_error.dot(layer_1.T)
    
    dlayer_1 = w4.T.dot(final_error)
    dlayer_1[layer_1<=0] = 0
    
    dw3 = dlayer_1.dot(fc.T)
    
    dfc = w3.T.dot(dlayer_1)
    dpool = dfc.reshape(pooled.shape)
    
    dconv_2 = pool_backprop(dpool,conv_2,pool_f,pool_s)
    dconv_2[conv_2<=0] = 0
    
    dfilt2, dconv_1  = conv_backprop(dconv_2,conv_1,filt2,conv_s)
    dconv_1[conv_1<=0] = 0
    
    dfilt1, dimage  = conv_backprop(dconv_1,image,filt1,conv_s)
    gradients = [dfilt1,dfilt2,dw3,dw4]
    
    return loss,gradients
```

```python
def optimizer(X,y,num_classes,lr,beta1,beta2,param,cost):
    [filt1,filt2,w3,w4] = param
    cost_per_batch = 0
    batch_size = len(X)
    
    # Gradients, momentum, RMSprop
    dfilt1 = np.zeros(filt1.shape)
    dfilt2 = np.zeros(filt2.shape)
    dw3 = np.zeros(w3.shape)
    dw4 = np.zeros(w4.shape)
    
    m1 = np.zeros(filt1.shape)
    m2 = np.zeros(filt2.shape)
    m3 = np.zeros(w3.shape)
    m4 = np.zeros(w4.shape)
    
    r1 = np.zeros(filt1.shape)
    r2 = np.zeros(filt2.shape)
    r3 = np.zeros(w3.shape)
    r4 = np.zeros(w4.shape)
    
    for i in range(batch_size):
        
        x = X[i]
        target = np.eye(num_classes)[y[i]].reshape(num_classes,1)
        
        loss,gradient = Implementation(x,target,param,1,2,2)
        [ddfilt1,ddfilt2,ddw3,ddw4] = gradient
        
        dfilt1 += ddfilt1
        dfilt2 += ddfilt2
        dw3 += ddw3
        dw4 += ddw4
        
        cost_per_batch += loss
        
    m1 = beta1*m1 + (1-beta1)*dfilt1/batch_size
    r1 = beta2*r1 + (1-beta2)*(dfilt1/batch_size)**2
    filt1 -= lr * m1/np.sqrt(r1 + 1e-8)
        
    m2 = beta1*m2 + (1-beta1)*dfilt2/batch_size
    r2 = beta2*r2 + (1-beta2)*(dfilt2/batch_size)**2
    filt2 -= lr * m2/np.sqrt(r2 + 1e-8)
        
    m3 = beta1*m3 + (1-beta1)*dw3/batch_size
    r3 = beta2*r3 + (1-beta2)*(dw3/batch_size)**2
    w3 -= lr * m3/np.sqrt(r3 + 1e-8)
        
    m4 = beta1*m4 + (1-beta1)*dw4/batch_size
    r4 = beta2*r4 + (1-beta2)*(dw4/batch_size)**2
    w4 -= lr * m4/np.sqrt(r4 + 1e-8)
    
    average_cost = cost_per_batch/batch_size
    cost.append(average_cost)
    
    param = [filt1,filt2,w3,w4]
    
    return param,cost
```

```python
def training(trainloader, num_classes = 10, beta1 = .9, beta2 = .999, img_depth = 3, f_dim = 5, lr = 0.01, num_filt1 = 8, num_filt2 = 16, num_epochs = 2, save_path = 'model.pkl'):
    
    filt1 = (num_filt1,img_depth,f_dim,f_dim)
    filt2 = (num_filt2,num_filt1,f_dim,f_dim)
    w3 = (128,2304)
    w4 = (10,128)
    
    filt1 = InitializeFilter(filt1)
    filt2 = InitializeFilter(filt2)
    w3 = InitializeWeight(w3)
    w4 = InitializeWeight(w4)
    
    param = [filt1,filt2,w3,w4]
    cost = []
    
    for e in range(num_epochs):
        
        for batch_i,(images,labels) in enumerate(trainloader):
            images,labels = images.numpy(),labels.numpy()
            
            param,cost = optimizer(images,labels,num_classes,lr,beta1,beta2,param,cost)
            
            if batch_i % 10 == 0:
                print('Epoch : {}\tTrainingloss = {: .4f}'.format(e+1,cost[-1]))
    
    with open(save_path,'wb') as f:
        pickle.dump(param,f)
        
```

```python
training(trainloader,10,.9,.999,3,5,.01,8,16,2)
```

```python
with open('model.pkl','rb') as f:
    model_param = pickle.load(f)
```

```python
test_loss = 0.0
class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))
batch_size = 20
num_classes = 10
for images,labels in testloader:
    
    images,labels = images.numpy(),labels.numpy()
    
    for i in range(batch_size):
        
        target = np.eye(num_classes)[labels[i]].reshape(num_classes,1)
        output,_ = ForwardPass(images[i],model_param,1,2,2)
        loss = CrossEntropyLoss(output,target)
        test_loss +=  loss
        max_in_output = np.argmax(output)
        checker = np.zeros((num_classes,1))
        checker[max_in_output] = 1
        boolean = np.array_equal(checker,target)
        
        if boolean:
            class_correct[max_in_output] += 1
            class_total[max_in_output] += 1
        else:
            correct_index = np.argmax(target)
            class_total[correct_index] += 1
        
avg_test_loss = test_loss/len(testloader.dataset)
print('Test Loss: {:.4f}\n'.format(avg_test_loss))

for i in range(num_classes):
    if class_total[i] > 0:
        print('Test Accuracy of %5s: %2d%% (%2d/%2d)' % (
            classes[i], 100 * class_correct[i] / class_total[i],
            np.sum(class_correct[i]), np.sum(class_total[i])))
    else:
        print('Test Accuracy of %5s: N/A (no training examples)' % (classes[i]))

print('\nTest Accuracy (Overall): %2d%% (%2d/%2d)' % (
    100. * np.sum(class_correct) / np.sum(class_total),
    np.sum(class_correct), np.sum(class_total)))   
```
